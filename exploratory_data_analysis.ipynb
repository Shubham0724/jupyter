```python
#univariate analysis: continues data ((scatterplot))
import pandas as pd
import seaborn as sns

df = pd.read_csv('iris.csv')
df
```

```python
import seaborn as sns
x_axis = df.index
y_axis = df['petal.length']
sns.scatterplot(x=x_axis, y=y_axis,hue=df['variety'])
```

```python
# strip plot
sns.stripplot(x=df['variety'], y=df['petal.length'])
```

```python
#distribution plot
sns.distplot(df['petal.length'])
```

```python
#histogram
import matplotlib.pyplot as plt
plt.hist(df['petal.length'])
```

```python
#box plot
sns.boxplot(x=df["petal.length"])
```

```python
#univariate analysis - categorical data ((count plot))
sns.countplot(x=df['variety'])
```

```python
#pie chart
_labels = ['SETOSA','VERSICOLOR','VIRGINICA']
plt.pie(df['variety'].value_counts(), labels=_labels)
```

```python
#bivariate analysis : continuous and continuous  ((scatterplot))
import pandas as pd
df= pd.read_csv('titanic.csv')
df
```

```python
sns.scatterplot(x=df['Fare'],y=df['Age'])
plt.title('Bivariate analysis')
```

```python
#corrleation
df[['Fare' , 'Age']].corr()
```

```python
# heatmap
sns.heatmap(df[['Fare','Age']].corr())
```

```python
#bivariate analysis  (categorical & categorical)
survived_ratio = df[['Pclass' , 'Survived']].groupby('Pclass').sum()
survived_ratio
```

```python
#bar plot
sns.barplot(x=survived_ratio.index, y=survived_ratio['Survived'])
```

```python
#bivariate analysis  :- continuous and categorical
sns.boxplot(x=df['Survived'],y=df['Age'])
```

```python
#Detecting Outliers
import pandas as pd
x=pd.Series([1,2,3,6,7,8,150])
std=x.std()
mean= x.mean()
z_scores = abs((x*mean)/std)
z_scores
```

```python
outliers_removed = x[z_scores <= 1.5]
outliers_removed
```

```python
import numpy as np
import pandas as pd
x= pd.Series([1,2,3,6,7,8,150])
mean=x.mean()
std=x.std()
median = np.median(x)
z_scores=abs((x-mean)/std)
median
```

```python
x[z_scores > 1.5] = median
x
```
